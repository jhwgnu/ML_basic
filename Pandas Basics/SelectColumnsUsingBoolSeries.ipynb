## Select columns composed with more than 80% of NaN

#### 1. Create Series of bool values

```python
lc0711.isnull().head()
```

```python
lc0711.isnull().sum().head()
```

```python
lc0711.isnull().sum()/len(lc0711) # (Series) +-/x (some value) = (Series)
```

```python
lc0711.isnull().sum()/len(lc0711)>=0.8 # (Series) (logical operation) = (bool Series)
```

#### 2. creating column index list using bool series

```python
bool_series = lc0711.isnull().sum()/len(lc0711)>=0.8
bool_series
```

```python
NaN_columns = lc0711.columns[bool_series]
NaN_columns
```
