### 두 데이터 병합하기 (option: how='outer' , 'inner' / on='key_id' /index='TRUE')

```python
## 임의의 데이터 셋 df1, df2 생성
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df2 = DataFrame({'key': ['a', 'b', 'd'],
                 'data2': ['x','y','z']})
df1
```

```python
# 생성한 df2의 data 구조 확인하기
df2
```

```python
# df1과 df2 병합하기 (일반적으로 option이 따로 없을 경우 데이터간 동일한 칼럼을 기준 키로 병합됨. 
# 동일한 키가 없는 경우 병합된 데이터에서는 사라짐)
pd.merge(df1,df2)
```

```python
# 위의 결과와 df1과 df2를 key id 기준으로 병합한 결과 비교하기
# on 옵션으로 동일한 변수를 키로 지정해주는것과 같은 결과
pd.merge(df1,df2,on='key')
```

```python
# 임의의 데이터 셋 df3, df4 생성
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'],
                 'data2': range(3)})
df3
```

```python
df4
```

```python
# 병합하려는 두 데이터에 동일한 변수명을 가진 key id가 없을 경우, 두 개의 데이터 셋에서 key id를 각각 지정해 줄 수 있음.
# df3 데이터에서는 lkey, df4 데이터에서는 rkey를 기준으로 병합하기
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
# 새로운 df1 ,  df2 생성하기
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                 'data1': range(6)})
df2 = DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                 'data2': ['AA','BB','CC','DD','EE']})
df1
```

```python
df2
```

```python
#두개의 데이터셋 df1과 df2를 'key'라는 변수를 기준으로 left 병합하기
pd.merge(df1, df2, on='key', how='left')
```

```python
#두개의 데이터셋 df1과 df2를 'key'라는 변수를 기준으로 inner 병합하기 (일종의 교집합 개념)
pd.merge(df1, df2, how='inner')
```

```python
##병합 기준 키가 되는 변수 외에 동일한 이름의 변수가 있는 경우

# 새로운 left ,  right 생성하기
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
left
```

```python
right
```

```python
#두개의 데이터셋 left와 right를 'key1'이라는 변수를 기준으로 병합하기
pd.merge(left, right, on='key1')
```

```python
#두개의 데이터셋 left와 right를 'key1'이라는 변수를 기준으로 병합하고 중복되는 변수명 다르게 지정해주기
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

```python
#두개의 데이터셋 left와 right를 'inner 병합하기
pd.merge(left, right, how='inner')
```

```python
# left1, right1 데이터 생성하기
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                  'value': range(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1
```

```python
right1
```

```python
# 두개의 데이터셋 left1과 right1을 left1에 있는 'key'라는 변수와 right1의 index 변수 기준으로 병합하기 
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
# 두개의 데이터셋 left1과 right1을 left1에 있는 'key'라는 변수와 right1의 index 변수 기준으로 outer 방식으로 병합하기 (일종의 합집합 개념
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

## 더미변수 생성하기

```python
#df 데이터 생성 후 범주형 변수인 key 변수에 대한 더미 변수 생성.
df = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                'data1': range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

```python
#더미 변수 명에 key라는 prefix 붙여서 dummies로 지정한 뒤, 기존 df 데이터와 join하여 df_with_dummy 데이터로 생성.
dummies = pd.get_dummies(df['key'], prefix='key')
```

```python
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
```

## 추가학습: 특이값 찾고 제거하기

```python
# 임의 1000 by 4 data 생성
np.random.seed(12345)
data = DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
#위에서 생성한 data의 한 칼럼에서 절대값이 3을 초과하는 값을 찾음. 
#이 경우 위의 분포에 따라 절대값 3 초과여부를 outlier 선정 기준으로 정함.
col = data[3]
col[np.abs(col) > 3]
```

```python
# any method를 이용하여 절대값 3을 초과하는 값이 들어있는 모든 행을 선택.
data[(np.abs(data) > 3).any(1)]
```

```python
# +3 초과 또는 -3 미만인 값들을 outlier로 간주하고 행 삭제.   
data[np.abs(data) > 3] = np.sign(data)*3
data.describe()
```
