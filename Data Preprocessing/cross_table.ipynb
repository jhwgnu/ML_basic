```python
import pandas as pd
df = pd.DataFrame(pd.read_csv("LoanStats3a.csv", engine='python'))
```

```python
pd.crosstab(df.grade, df.term)
```

```python
pd.crosstab(df.grade, df.term, margins=True)
```

```python
pd.crosstab(df.grade, df.term, margins=True, normalize=True)
```
