```python
import pandas as pd
import numpy as np
df = pd.DataFrame(pd.read_csv('LoanStats3a.csv', engine='python'))
```

## 1. Creating a Y/N variable

```python
df.loan_status.head()
```

```python
df.loan_status.value_counts()
```

```python
df['default'] = np.where(df.loan_status=='Fully Paid', "N", "Y")
df[['loan_status', 'default']].head(10)
```

```python
df.default.value_counts()
```

## 2. Merging various variables using np.where

```python
df.purpose.value_counts()
```

```python
df['purpose1'] = np.where((df.purpose=="debt_consolidation")|(df.purpose=="credit_card"), df.purpose, 'others')
```
